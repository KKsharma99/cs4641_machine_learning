#### Plotting Randomized Hill Climbing

```python
with open('output/rhc_results.csv') as f:
    rhc_str = f.readlines()
print(rhc_str)

rhc = []
for i in range(len(rhc_str)):
    rhc.append(list(map(float, rhc_str[i].strip('\n').strip(',').split(','))))

rhc = np.array(rhc)

plt.figure()
plt.plot(range(11),rhc[0,:11], range(11), rhc[2,:11])
plt.ylabel('Error')
plt.xlabel('Iterations')
plt.xticks(range(12), ['100', '200', '300', '400', '500', '700', '900', '1000', '1200', '1500', '1700'])
#plt.yticks(range(12), [5, 10, 15, 20, 25, 30, 35, 40, 45, 50, 55, 60])
plt.ylim(top = 35) 
plt.title('Error vs iterations curve for Randomized hill climbing')
plt.legend(['Training Error','Testing Error'])
plt.show()
```

#### Plotting Simulated Annaeling

```python
cool_exp = [0.100, 0.200, 0.300, 0.400, 0.500, 0.600, 0.700, 0.800, 0.900]

sa = []
for ce in cool_exp:

    filename = 'output/' + 'sa_'+str(ce)+'00_results.csv'
    with open(filename) as f:
        data_str = f.readlines()

    data = []
    for i in range(len(data_str)):
        data.append(list(map(float, data_str[i].strip('\n').strip(',').split(','))))

    sa.append(np.array(data))
```

```python
#Training Error
fig2, ax = plt.subplots()
for index, ce in enumerate(cool_exp):
    if(ce == 0.2):
        lw = 3
    else:
        lw = 1
    ax.plot(range(12), sa[index][0,:], label =str(ce), linewidth= lw)
plt.xticks(range(12),['100', '200', '300', '400', '500', '700', '900', '1000', '1200', '1500', '1700', '2000'])
ax.legend()
plt.ylabel('Training Error')
plt.xlabel('Iterations')
plt.title('Training Error vs Iterations, Simulated Annealing')
plt.show()
```

```python
#Testing Error
fig3, ax2 = plt.subplots()
for index, ce in enumerate(cool_exp):
    if(ce == 0.2):
        lw = 3
    else:
        lw = 1
    ax2.plot(range(12), sa[index][2,:], label =str(ce), linewidth = lw)
plt.xticks(range(12),['100', '200', '300', '400', '500', '700', '900', '1000', '1200', '1500', '1700', '2000'])
ax2.legend()
plt.ylabel('Testing Error')
plt.xlabel('Iterations')
plt.title('Testing Error vs Iterations, Simulated Annealing')
plt.show()
```

#### Plotting Genetic Algorithm

```python
pop = [10, 50, 100]
mate = [5, 10, 25]
mutate = [5, 10, 25]
ga = []
for index, p in enumerate(pop):

    filename = 'output/' + 'ga_'+str(p)+'.000_'+str(mate[index])+'.000_'+str(mutate[index])+'.000_'+'results.csv'

    print(filename)
    with open(filename) as f:
        data_str = f.readlines()

    data = []
    for i in range(len(data_str)):
        data.append(list(map(float, data_str[i].strip('\n').strip(',').split(','))))

    ga.append(np.array(data))
```

```python
#Training Error
fig4, ax3 = plt.subplots()
for index, p in enumerate(pop):
    if(index == 2):
        lw = 3
    else:
        lw = 1

    label = str(p)+','+str(mate[index])+','+str(mutate[index])
    ax3.plot(range(16), ga[index][0,:], label = label, linewidth = lw)

plt.xticks(range(16),['100', '200', '300', '400', '500', '700', '900', '1000', '1200', '1500', '1700', '2000', '4000', '8000', '16000', '20000'], rotation = 45)
ax3.legend()
plt.ylabel('Training Error')
plt.xlabel('Iterations')
plt.title('Training Error vs iterations curve for Genetic Algorithm')
plt.show()
```

```python
#Testing Error
fig5, ax4 = plt.subplots()
for index, p in enumerate(pop):
    label = str(p)+','+str(mate[index])+','+str(mutate[index])

    if(index == 2):
        lw = 3
    else:
        lw = 1

    ax4.plot(range(16), ga[index][2,:], label = label, linewidth= lw)

    plt.xticks(range(16),['100', '200', '300', '400', '500', '700', '900', '1000', '1200', '1500', '1700', '2000', '4000','8000','16000','20000'], rotation = 45)
ax4.legend()
plt.ylabel('Testing Error')
plt.xlabel('Iterations')
plt.title('Testing Error vs iterations curve for Genetic Algorithm')
plt.show()
```

```python
time_elapsed = [49.043, 113.424, 263.335]
x_ticks = []
for i in pop:
    x_ticks.append(str(i))
plt.plot( x_ticks, time_elapsed )
plt.xlabel('Population')
plt.ylabel('Training Time')
plt.title('Training Time for Various Population Sizes')
plt.show()
```

```python
final_error = [26.56, 23.958, 21.354 ]
x_ticks = []
for i in pop:
    x_ticks.append(str(i))
plt.bar( x_ticks, final_error )
plt.xlabel('Cooling Rate')
plt.ylabel('Final Error Rate')
plt.title('Final Error Rate for Various Population Sizes')
plt.show()
```

#### Plotting Travelling Salesman Problem

```python
#Plotting the optimal evaluation function

with open('output/' + 'travelling_salesman_test.csv') as f:
    tsp_str = f.readlines()

tsp = []
for i in range(len(tsp_str) - 1):
    tsp.append(list(map(float, tsp_str[i].strip('\n').strip(',').split(','))))

tsp = np.array(tsp)

fig1, ax = plt.subplots()
labels = ['RHC','SA', 'GA']

for i in list(range(3)):
    ax.plot(range(9), tsp[i,:], label = labels[i])

plt.xlabel('Iterations')
plt.ylabel('Best fitness value')
plt.legend()
plt.xticks(range(9),['50', '100', '200', '300', '500', '700', '1000', '5000', '10000'])
plt.title('Optimal Fitness Values, Traveling Salesman')
plt.show()
```

```python
#Plotting the time taken for each algorithm for varying iterations

fig2, ax = plt.subplots()

l = [4,5,6]
for i in l:
    ax.plot(range(9), tsp[i,:], label = labels[i-4])

plt.xlabel('Iterations')
plt.ylabel('Total time taken')
plt.legend()
plt.xticks(range(9), ['50', '100', '200', '300', '500', '700', '1000', '5000', '10000'])
plt.title('Total time taken with varying iterations (travelling salesman)')
plt.show()
```

#### Plotting Travelling Salesman Problem - Genetic Algorithm Results

```python
#Plotting the evaluation function value

with open('output/' + 'tsp_ga.csv') as f:
    tsp_str = f.readlines()

tsp = []
for i in range(len(tsp_str)):
    tsp.append(list(map(float, tsp_str[i].strip('\n').strip(',').split(','))))

tsp = np.array(tsp)
fig, ax = plt.subplots()
ax.bar(range(np.shape(tsp)[1]), tsp[0,:])
plt.xlabel("Parameters")
plt.ylabel("Optimal fitness values")
plt.title("GA Fitness Values for Various Hyper Parameters")
plt.xticks(range(np.shape(tsp)[1]), ["10, 5, 5","20, 10, 10","50, 25, 25","100, 50, 50","200, 100, 100"], rotation=20)

plt.show()
```
